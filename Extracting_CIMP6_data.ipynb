```python
from netCDF4 import Dataset 
import pandas as pd
import numpy as np

data = Dataset('tos_Omon_CanESM5_historical_r1i1p1f1_gn_19800116-20141216_v20190429.nc', 'r')
#time_var = data.variables['time']
data.variables

df = pd.DataFrame(0.0, columns = ['t_min'], index = [i for i in range(12418)])
locations = pd.read_excel("Uganda.xlsx")
#getting the name, lon, and lat  of the locations 
for index, row in locations.iterrows():
    location = row['Station']
    location_latitude = row['Latitude']
    location_longitude = row['Longitude']
    # reading the data using the fnames list
    
    data = Dataset('tasmin_day_NESM3_historical_r1i1p1f1_gn_19810101-20141231_v20190812.nc', 'r')

    # Storing the lat and lon data of the netCDF file into variables 
    lat = data.variables['lat'][:]
    lon = data.variables['lon'][:]

    # Squared difference between the specified lat,lon and the lat,lon of the netCDF 
    sq_diff_lat = (lat - location_latitude)**2 
    sq_diff_lon = (lon - location_longitude)**2

    # Identify the index of the min value for lat and lon
    min_index_lat = sq_diff_lat.argmin()
    min_index_lon = sq_diff_lon.argmin()

    # Accessing the average precipitation data
    pr = data.variables['tasmin']
    #writing the extracted precipitation data to the dataframe
    df['t_min'] = pr[:, min_index_lat, min_index_lon]
    df['t_min'] = df['t_min'] - 273.15

df.to_excel(location+".xlsx")
```
